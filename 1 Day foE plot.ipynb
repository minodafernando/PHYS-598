```python
import datetime

data_timestring = []
data_foE = []

filename = "foE.txt"
fp = open(filename, 'r')
for line in fp:
    #print(line)
    if(line[0] == '#'):
        continue
    else:
        line = line.strip()
        line_split = line.split()
        data_foE.append(float(line_split[2]))
        this_ts = datetime.datetime.strptime(line_split[0], "%Y-%m-%dT%H:%M:%S.%fZ")
        data_timestring.append(this_ts)

print(data_timestring)
print("\n")
print(data_foE)

fp.close()
```

#### String to datetime conversion

##### %Y: 4-digit year
##### %y: 2-digit year
##### %m: 2-digit month (01-12)
##### %d: 2-digit day of the month (01-31)
##### %H: 2-digit hour (00-23)
##### %M: 2-digit minute (00-59)
##### %S: 2-digit second (00-59)

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(18, 10))
ax.plot(data_timestring, data_foE)
plt.xlabel("Time", fontsize = 20)
plt.ylabel("foE Data", fontsize = 20)
plt.show()
```
