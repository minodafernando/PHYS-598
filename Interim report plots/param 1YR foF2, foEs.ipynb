## 1 Year (2012/01/01-2013/01/01) foEs and foF2

### Read file

```python
## Import stuff
import datetime
import numpy as np

# Define timestamp array
timestamp_full = []

# Define CS array
CS = []

# F2 layer critical frequency
foF2 =[]
timestamp_foF2 = []
# Es layer critical frequency
foEs = []
timestamp_foEs = []


# Define filename
filename = "(edited) GD 01-01-2012 00H01M to 01-01-2013 00H00M.txt"

# define indices for each parameter of file
time_idx = 0
CS_idx = 1
foF2_idx = 2
foEs_idx = 4

# Define format for datetime
format = "%Y-%m-%dT%H:%M:%S.%fZ"

# Open file - read
fp = open(filename, 'r')
# Read every line in file
for line in fp:
# Skip comments
    if line[0] == "#":
        continue
# Strip lines
    else:
        line_strip = line.strip()
# Split lines into lists
        line_split = line.split()

        foF2_str = line_split[foF2_idx]
        foEs_str = line_split[foEs_idx]

        timestamp_full.append(datetime.datetime.strptime(line_split[time_idx], format))
        
        # Check if can convert foE to float
        try:
            foF2_float = float(foF2_str)
            # If successful, append to the array
            foF2.append(foF2_float)
            # Extract timestamp and append to the array
            timestamp_str = line_split[time_idx]
            timestamp_foF2.append(datetime.datetime.strptime(timestamp_str, format))
        except ValueError:
            # If the conversion fails, ignore the line
            pass

        # Check if can convert foEs to float
        try:
            foEs_float = float(foEs_str)
            # If successful, append to the array
            foEs.append(foEs_float)
            # Extract timestamp and append to the array
            timestamp_str = line_split[time_idx]
            timestamp_foEs.append(datetime.datetime.strptime(timestamp_str, format))
        except ValueError:
            # If the conversion fails, ignore the line
            pass




#print("Number of hE height data lines:", len(hE))
print("Number of foF2 data lines:", len(foF2))
print("Number of timestamp_foE data lines:", len(timestamp_foF2))
print("Number of foEs data lines:", len(foEs))
print("Number of timestamp_foEs data lines:", len(timestamp_foEs))
print("\n")
print("check consistency:", "\n")
print("foF2[1]: ", foF2[1], timestamp_foF2[1])
print("foEs[1]: ", foEs[1], timestamp_foEs[1])
print("\n")
#print("Lowest height found:", min(hE))
print("Type of data of hE:", type(foF2[1]))

# Close file
fp.close()
```

### foF2 vs time - plot

```python
import matplotlib.pyplot as plt

plt.scatter(timestamp_foF2, foF2, s=2, color="midnightblue")
plt.title("foF2 in MHz vs Time in YYYY-MM for one year (2012/01/01 to 2012/12/31)", pad=15)
plt.xlabel("Time (YYYY-MM)", labelpad=10)
plt.ylabel("foF2 (MHz)", labelpad=6)

plt.xticks(rotation=90)
plt.show()
```

### foEs vs time - plot

```python
import matplotlib.pyplot as plt

plt.scatter(timestamp_foEs, foEs, s=2, color="darkgoldenrod")
plt.title("foEs in MHz vs Time in YYYY-MM for one year (2012/01/01 to 2012/12/31)", pad=15)
plt.xlabel("Time (YYYY-MM)", labelpad = 10)
plt.ylabel("foEs (MHz)", labelpad = 10)

plt.xticks(rotation=90)
plt.show()
```
