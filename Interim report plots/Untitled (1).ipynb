```python
import matplotlib.pyplot as plt
import numpy as np

# Create an array of angles from 0 to 2*pi
theta = np.linspace(0, 2*np.pi, 100)

# Create points in the annulus
z_annulus = 2 + 2*np.exp(1j*theta)

# Plot the annulus
plt.plot(np.real(z_annulus), np.imag(z_annulus), label='Annulus: 1 < |z-2| < 3')

# Plot the mapped points under f(z) = z
plt.scatter([2, 2, 3, 1], [1, -1, 0, 0], color='red', label='Mapped points')

plt.title('Mapping of points in the annulus under f(z) = z')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()
```
