```python
import matplotlib.pyplot as plt

# Sample data (replace with your actual data)
hours = [1, 2, 3, 4, 5]  # Hour of the day
hEs_points = [10, 15, 20, 25, 30]  # Number of hEs points per hour
riometer_absorption = [5, 10, 15, 20, 25]  # Average/peak riometer absorption

# Plotting the data
plt.figure(figsize=(8, 6))
plt.scatter(riometer_absorption, hEs_points, c=hours, cmap='viridis', label='Hour')
plt.colorbar(label='Hour of the day')
plt.xlabel('Average/Peak Riometer Absorption')
plt.ylabel('Number of hEs Points')
plt.title('hEs Points per Hour vs. Riometer Absorption')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data (replace with your actual data)
timestamps = ["2024-03-06 01:30:00", "2024-03-06 02:15:00", "2024-03-06 03:40:00", "2024-03-06 04:20:00", "2024-03-06 01:10:00"]
hEs_points = [10, 15, 20, 25, 30]  # Number of hEs points per hour
riometer_absorption = [5, 10, 15, 20, 25]  # Average/peak riometer absorption

# Convert timestamps to datetime objects
timestamps = pd.to_datetime(timestamps)

# Categorize timestamps into hours
hours = timestamps.hour

# Plotting the data
plt.figure(figsize=(8, 6))
plt.scatter(riometer_absorption, hEs_points, c=hours, cmap='viridis', label='Hour')
plt.colorbar(label='Hour of the day')
plt.xlabel('Average/Peak Riometer Absorption')
plt.ylabel('Number of hEs Points')
plt.title('hEs Points per Hour vs. Riometer Absorption')
plt.grid(True)
plt.legend()
plt.show()
```
