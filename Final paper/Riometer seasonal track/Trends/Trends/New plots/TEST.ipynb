```python
import matplotlib.pyplot as plt

def sum_list(lst):
    new_list = []
    for element in lst:
        new_list.append(element + 1)
    return new_list

A = [1, 3, 4, 5]
B = [4, 3, 5, 1]
C = [3, 6, 2, 2]

# Create a dictionary to store the new lists
new_lists = {}

# Loop through each old list
for name, old_list in [('A', A), ('B', B), ('C', C)]:
    # Compute the new list
    new_lists['new_list' + name] = sum_list(old_list)

# Define the time list
time = [0, 10, 20, 30]

# Plotting
plt.figure(figsize=(8, 6))  # Adjust figure size if needed
for key, value in new_lists.items():
    plt.plot(time, value, label=key)  # Plot each new list against time

plt.xlabel('Time')
plt.ylabel('Value')
plt.title('New Lists Plotted Against Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example data
hEs_values = [10, 20, 30, 40, 50]
absorption_values = [1, 2, 3, 4, 5]

# Calculate the standard deviation for each absorption value
std_deviation_dict = {}

for i, absorption in enumerate(absorption_values):
    hEs_subset = [hEs_values[j] for j, abs_value in enumerate(absorption_values) if abs_value == absorption]
    std_deviation_dict[absorption] = np.std(hEs_subset)

# Plot hEs vs absorption
plt.plot(absorption_values, hEs_values, marker='o', label='hEs vs Absorption')

# Plot standard deviation against absorption
std_deviation_values = [std_deviation_dict[abs_value] for abs_value in absorption_values]
plt.errorbar(absorption_values, hEs_values, yerr=std_deviation_values, fmt='none', ecolor='red', capsize=5, label='Standard Deviation')

plt.xlabel('Absorption')
plt.ylabel('hEs')
plt.title('hEs vs Absorption with Standard Deviation')
plt.legend()
plt.grid(True)
plt.show()
```
